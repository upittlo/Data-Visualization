```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
reviews =pd.read_csv('data/winemag-data_first150k.csv',index_col = 0)
reviews.head()
```

### Style the plot

- figsize = (width,height)
- color = 'xxx'
- fontsize
- title


```python
reviews['points'].value_counts().sort_index().plot.bar(figsize = (12,6),
                                                      color = 'mediumvioletred',
                                                       fontsize = 16,
                                                      title = 'Ranking Given by Wine Magazine')
```

### Use Matplotlib for deeper control

- set_title("title content",fontsize)
- sns.despine(bottom = True,left = True)


```python
ax = reviews['points'].value_counts().sort_index().plot.bar(
     figsize = (12,6),
     color = 'mediumvioletred',
     fontsize = 16
)

ax.set_title("Ranking Given Wine Magazine",fontsize = 20)
sns.despine(bottom = True,left = True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pokemon = pd.read_csv("data/Pokemon.csv")
pokemon.head(3)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pokemon = pd.read_csv("data/Pokemon.csv")
pokemon.head(3)
```

```python
pokemon.plot.scatter(x = 'Attack',y = 'Defense',
                    title = 'Pokemon by Attack and Defense',
                    figsize = (12,6))
```

```python
ax = pokemon['Total'].plot.hist(
figsize = (12,6),
fontsize =14,
bins = 50,
color = 'gray')

ax.set_title('Pokemon by Stat Total',fontsize = 20)
```

```python
ax = pokemon['Type 1'].value_counts().plot.bar(
figsize = (12,6), color = 'blue',
fontsize = 16)

ax.set_title("Pokemon by Primary Type",fontsize = 20)
sns.despine(bottom = True,left = True)
```
