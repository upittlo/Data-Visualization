```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
reviews = pd.read_csv("data/winemag-data_first150k.csv",index_col = 0)
reviews.head(3)
```

### Bar Charts and categorical data

- Mostly for nominal categorical data
- value_counts(): 計算類別個數
- pandas.plot.bar() :Bar Chart


```python
reviews['province'].value_counts().head(10).plot.bar()
```

#### Plot ratio

- Pandas series can be divided

```python
(reviews['province'].value_counts().head(10)/len(reviews)).plot.bar()
```

```python
reviews['points'].value_counts().sort_index().plot.bar()
```

### Line Chart

- Mostly for ordinary categorical data
- Use for continuous values
- A line chart can pass over any number of many individual values, making it the tool of first choice for distributions with many unique values or categories.

- weakness: unlike bar charts, they're not appropriate for nominal categorical data. 


```python
reviews['points'].value_counts().sort_index().plot.line()
```

### Area Chart



```python
reviews['points'].value_counts().sort_index().plot.area()
```

### Histograms

- Interval data
- a histogram is special kind of bar plot that splits your data into even intervals and displays how many rows are in each interval with bars. The only analytical difference is that instead of each bar representing a single value, it represents a range of values.

- Histograms work best for interval variables without skew. They also work really well for ordinal categorical variables

#### Not good for scale data >>Solution: cut off at a sensible level

```python
reviews[reviews['price']<200]['price'].plot.hist()
```
