```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('max_columns', None)
df = pd.read_csv("data/CompleteDataset.csv", index_col=0)

import re
import numpy as np

footballers = df.copy()
footballers['Unit'] = df['Value'].str[-1]
footballers['Value (M)'] = np.where(footballers['Unit'] == '0', 0, 
                                    footballers['Value'].str[1:-1].replace(r'[a-zA-Z]',''))
footballers['Value (M)'] = footballers['Value (M)'].astype(float)
footballers['Value (M)'] = np.where(footballers['Unit'] == 'M', 
                                    footballers['Value (M)'], 
                                    footballers['Value (M)']/1000)
footballers = footballers.assign(Value=footballers['Value (M)'],
                                 Position=footballers['Preferred Positions'].str.split().str[0])
```

```python
footballers.head()
```

### Scatter Plot

- seeing which type of offensive players tends to get paid the most: the striker, the right-winger, or the left-winger.

- "hue"  for color

```python
import seaborn as sns

sns.lmplot(x = 'Value',y = 'Overall',hue = 'Position',
          data = footballers.loc[footballers['Position'].isin(['ST','RW','LW'])],
                                fit_reg = False)
```

#### Add marker for different position

```python
sns.lmplot(x = 'Value', y= 'Overall',markers = ['o','x','*'],hue = 'Position',
          data = footballers.loc[footballers['Position'].isin(['ST','RW','LW'])],
                                fit_reg = False)
```

#### Grouped box plot

- Grouping is an extremely communicative visual variable: it makes this chart very easy to interpret. However, it has very low cardinality: it's very hard to use groups to fit more than a handful of categorical values. 

```python
f=  (footballers.loc[footballers['Position'].isin(['ST','GK'])]
    .loc[:,['Value','Overall','Aggression','Position']])

f = f[f['Overall']>=80]
f = f[f["Overall"] <85]
f['Aggression'] = f["Aggression"].astype(float)

sns.boxplot(x = 'Overall',y = 'Aggression',hue = "Position",data = f)
```

### Summarization

- difficult to squeeze enough dimensions onto a plot without hurting its interpretability
- creation and addition of new variables by mixing and matching the information provided in the old ones
-  "boil down" potentially very complicated relationships into simpler ones.

#### Heatmap

- measures the correlation between every pair of values in a dataset and plots a result in color.

```python
f = (
footballers.loc[:,['Acceleration','Aggression','Agility','Balance','Ball control']]
    .applymap(lambda v:int(v) if str.isdecimal(v) else np.nan)
    .dropna()
).corr()

sns.heatmap(f,annot = True)
```
